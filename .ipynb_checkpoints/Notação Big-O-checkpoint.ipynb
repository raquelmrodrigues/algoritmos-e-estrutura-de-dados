# Função 1 - O(n)

```python
def soma1(n):
    soma = 0
    for i in range(n + 1):
        soma += i
        
    return soma
```

```python
soma1(10)
```

```python
%timeit soma1(10)
```

## Função 2 - O(3)

```python
def soma2(n):
    return (n * (n + 1)) / 2
```

```python
soma2(10)
```

```python
%timeit soma2(10)
```

## Função 3 - O(n)

```python
def lista1():
    lista = []
    for i in range(1000):
        lista += [i]
    return lista
```

```python
%timeit lista1()
```

# Funções Big-O

```python
from math import log
import numpy as np
import matplotlib.pyplot as plt

n = np.linspace(1, 10, 100)
```

```python
labels = ['Constant', 'Logarithmic', 'Linear', 'Log linear', 'Cubic', 'Exponential']
big_o = [np.ones(n.shape), np.log(n), n, n * np.log(n), n**2, n**3, 2**n]
```

```python
plt.figure(figsize=(10,8))
plt.ylim(0,100)
for i in range(len(big_o)):
    plt.plot(n, big_o[i], label = labels[i])
    
plt.legend()
plt.ylabel('Runtime')
plt.xxlabel("n")
```

# Exemplos Big-O

```python
lista = [1, 2, 3 , 4, 5]
```

## Constant - O(1)

```python
def constant(n):
    print(n[0])
```

```python
constant(lista)
```

## Linear - O(n)

```python
def linear(n):
    for i in n:
        print(i)
```

```python
linear(lista)
```

## Quadratic - O(n^2) - polynomial

```python
def quadratic(n):
    for i in n:
        for j in n:
            print(i, j)
        print('---')
```

```python
quadratic(lista)
```

## Combination

```python
def combination(n):
    # O(1)
    print(n[0])
    
    # O(5)
    for i in range(5):
        print('test', i)
        
    # O(n)
    for i in n:
        print(i)
    
    # O(n)
    for i in n:
        print(i)
    
    # O(3)
    print('Python')
    print('Python')
    print('Python')
    
    # total = O(9) + O(2n) -> O(n)
```

```python
combination(lista)
```
